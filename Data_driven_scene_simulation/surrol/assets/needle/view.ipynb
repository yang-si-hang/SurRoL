```python
import numpy as np
import matplotlib.pyplot as plt
sdf=np.load('./needle_sdf256.npy')
sdf.shape
```

```python
slice=sdf[:,:,127]
plt.imshow(1-slice,cmap='YlGnBu')
plt.axis('off')
plt.savefig('./sdf.png',dpi=144,bbox_inches="tight")
```

```python
threshold=0.01
idx=slice>threshold
plt.axis('off')
plt.imshow(1-idx,cmap='YlGnBu')
plt.savefig('./bool.png',dpi=144,bbox_inches='tight')
```
